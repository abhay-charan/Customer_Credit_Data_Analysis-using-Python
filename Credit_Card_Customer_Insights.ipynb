```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("credit_card.csv")
df.head()
```

```python
df['Limit'] = df['Limit'].str.replace("INR ", "", regex=False)
df['Limit'] = df['Limit'].str.replace(",", "", regex=False)
df['Limit'] = df['Limit'].astype(float)
```

```python
df['Credit_Type'].value_counts()
```

```python
df['Segment'].value_counts()
```

```python
df.groupby('Segment')['Limit'].mean()
```

```python
df['Credit_Type'].value_counts().plot(kind='bar')
plt.title("Customers by Credit Type")
plt.xlabel("Credit Type")
plt.ylabel("Number of Customers")
plt.show()
```

```python
df['Segment'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Customers Distribution by Segment")
plt.ylabel("")
plt.show()
```

```python
plt.hist(df['Limit'], bins=10)
plt.title("Distribution of credit limit")
plt.xlabel("Credit Limit")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x='Segment', y='Age', data=df)
plt.title("Age Distribution by Segment")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
